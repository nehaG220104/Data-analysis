```python
import pandas as pd
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv('movie_dataset.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
feature=['keywords','cast','genres','director']
```

```python
def combine_feacture(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
```

```python
for features in feature:
    df[features]=df[features].fillna('')
df['combined_feature']= df.apply(combine_feacture,axis=1)
```

```python
cv= CountVectorizer()
```

```python
count_matrix=cv.fit_transform(df["combined_feature"])
```

```python
cos= cosine_similarity(count_matrix)
```

```python
def get_title_for_index(index):
    return df[df.index == index]["title"].values[0]
```

```python
def get_index_for_title(title):
    return df[df.title == title]["index"].values[0]
```

```python
user_like='Avatar'
index = get_index_for_title(user_like)
similar=list(enumerate(cos[index]))
```

```python

sorted_list = sorted(similar, key=lambda x:x[1],reverse=True)[1:]
```

```python
i=0
print("top 5 movies:"+user_like)
for element in sorted_list:
    print(get_title_for_index(element[0]))
    i=i+1
    if i>5:
        break
```
